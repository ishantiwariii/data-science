## here we are importing pandas

```python
import numpy as np
import pandas as pd
```

### series from list 


```python
#string 

country = ['india' , 'america' , 'australia' , 'united kingdom']

pd.Series(country)
```

```python
#integer 

num = [1, 2, 3, 4, 5, 6]

pd.Series(num)
```

### creating custum index 


```python
marks = [45, 65, 78, 89, 90]
subject = ["os" , "Ds" , "Digital" , "Mathematics" , "DBMS"]

pd.Series(marks , index = subject)
```

```python
#giving a name to my series 

marks = pd.Series(marks , index = subject , name = "ishan k marks")
marks
```

### by the use of dictionary 

```python
dict1={
  "maths" :98,
  "DBMS" :96,
  "OS" :84,
  "DS" :96
}

marks1= pd.Series(dict1 , name = "isha k marks")
marks1
```
