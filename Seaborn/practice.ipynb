```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

1. Relational Plots

- to see a statistical relation between 2 or more variable
- bivariate analysis

  plots under this section 
    - scatterplot
    - lineplot

```python
# it is instructed by the seaborn library that we should use the figure level functions

## sns.relplot()> sns.lineplot() / sns.scatter()
# in sns.relplot(kind='') we can provide the kind of the graph which can be line or whatever it is... 
```

```python
# facets are used when we are using figure level functions.....
# facets means breaking the graph on the category basis
# like rows and columns 
# so instead of hue we provide row= , col= 
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips, x='total_bill' , y='tip' , hue='sex' , style='time', size='size')
```

```python
sns.relplot(data=tips, x='total_bill' , y='tip', kind='scatter' , hue='sex' , style='time', size='size')
```

```python
# if we want continuity in our graph then we will use line plot

gap = px.data.gapminder()
gap
```

```python
ind= gap[gap['country']=='India']
```

```python
sns.lineplot(data=ind, x='year' , y='lifeExp')
```

```python
sns.relplot(data=ind, x='year' , y='lifeExp' , kind='line')
```

```python
temp = gap[gap['country'].isin(['India' , 'Germany' , 'Uganda'])]
temp
```

```python
sns.lineplot(data=temp , x='year' , y='lifeExp' , hue='country' , style='continent')
```

```python
sns.relplot( kind='line' ,data= temp , x='year' , y='lifeExp' , hue='country' , style='continent')
```

```python
tips
```

```python
# facet plots 

# if we want to divide the data of the plots on basis of categorical then we use col and row attributes

# again if we want to apply these in our plots we must have to use the figure function which is relplot

sns.relplot(kind='scatter', data=tips, x='tip' , y='total_bill' , col='sex')
```

```python
sns.relplot(kind='scatter', data=tips, x='tip' , y='total_bill' , row='sex', col='smoker')
```

```python
# col_wrap -> we also can wrap the columns as per our requirements 
sns.relplot(kind='scatter', data=gap , x='lifeExp' , y='gdpPercap' , col='year' , col_wrap=3)
```

2. Distribution plots 

  - used for univariate analysis
  - used to find out the distribution 
  - range of the observation 
  - central tendency 
  - is the data bimodal ?
  - are there outliers ?

plots under this 
- histplot
- KDEplot
- Rugplot

```python
# histogram 
# figure level function -> displot()
# axes level -> histplot -> kdeplot -> rugplot

tips
```

```python
sns.displot(data=tips, x='total_bill' , kind='hist')
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
# bins para
sns.displot(data=tips, x='total_bill' , kind='hist' , bins= 20)
```

```python
sns.displot(data=tips, x='tip' , kind='hist' , hue='sex' )
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age' , kind='hist' , hue='sex', element='step' )
```

```python
sns.displot(data=titanic, x='age' , kind='hist' , col='sex' , element='step' )
```

```python
# KDE -> kernal density estimation 
tips
sns.kdeplot(data=tips, x='total_bill', hue='sex' , fill=True)
```

```python
# hue -> fill 
sns.displot(kind='kde' , data=tips , x='total_bill' , hue='sex' , fill=True)
```

```python
# rugplot
sns.kdeplot( data=tips , x='total_bill' )
sns.rugplot(data=tips , x='total_bill')
```

```python
# bivariate histogram 
sns.histplot(data=tips , x='total_bill' , y='tip')
sns.displot(kind='hist',data=tips , x='total_bill' , y='tip')
```

```python
sns.displot(kind='kde', data=tips, x='total_bill' , y='tip')
```

```python
temp= gap[gap['continent']=='Asia'].pivot(index='country', columns='year' ,values='lifeExp')
plt.figure(figsize=(15, 10))
sns.heatmap(temp , annot=True , linewidths=0.5, cmap='viridis')
```
